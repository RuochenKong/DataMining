```python
import soundfile as sf
import numpy as np
import pandas as pd

data, samplerate = sf.read('..\\data\\train\\000000.ogg')
df = pd.DataFrame(data, columns = ['channel-1','channel-2'])
df
```

```python
import matplotlib.pyplot as plt
from PIL import Image

plt.figure(figsize = (40,6),dpi = 80)
plt.plot(df['channel-1'])
plt.title('Electronic Channel 1')
fn1 = 'Ex_Elec_Ch1.png'
plt.savefig(fn1)

plt.figure(figsize = (40,6),dpi = 80)
plt.plot(df['channel-2'])
plt.title('Electronic Channel 2')
fn2 = 'Ex_Elec_Ch2.png'
plt.savefig(fn2)

im1 = Image.open(fn1)
im2 = Image.open(fn2)
width,height = im1.size
result = Image.new(im1.mode,(width,height*2))
result.paste(im1,box=(0,0))
result.paste(im2,box=(0,height))
result.save('Electronic.png')
```
