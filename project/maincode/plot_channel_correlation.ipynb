```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '../data/corr/'
cAll = np.load(path + 'all.npy')
cAll = pd.Series(cAll)
plt.figure(figsize = (10,4),dpi = 80)
sns.boxplot(x = cAll, width = 0.3)
plt.show()
```

```python
values = []
for i in range(19):
    cor = np.load(path + '%02d.npy'%i)
    values.append(pd.Series(cor))

plt.figure(figsize = (20,6),dpi = 80)
sns.boxplot( data = values , width = 0.4);
plt.title('Correlation between channels')
plt.savefig('../figures/cor_channels.png')
```

```python
plt.figure(figsize = (20,6),dpi = 80)
sns.boxplot( data = values , width = 0.4, showfliers = False);
plt.title('Correlation between channels - without outliers')
plt.savefig('../figures/cor_channels_not_outliers.png')
```
